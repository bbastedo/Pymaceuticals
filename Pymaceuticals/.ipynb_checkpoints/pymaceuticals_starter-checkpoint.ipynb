```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_df = pd.merge(mouse_data, clinical_trial, on='Mouse ID')

# Display the data table for preview

merged_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_treatment_grouped = merged_df.groupby(("Drug","Timepoint"))
mean_tumor_vol = tumor_treatment_grouped['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
tumor_response_df = pd.DataFrame(mean_tumor_vol)
# Preview DataFrame
tumor_response_df = tumor_response_df.reset_index()
timepoint = tumor_response_df['Timepoint']
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
std_error_of_tumors = merged_df.groupby(("Drug","Timepoint"))
std_error = std_error_of_tumors['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
std_error_df = pd.DataFrame(std_error).reset_index()
volume = tumor_response_df['Tumor Volume (mm3)']
# Preview DataFrame
std_error_df.head()
```

```python
# Minor to Re-Format the Data Frames
transpose_df_tumor_response = tumor_response_df.unstack(1).transpose()
transpose_df_std_error = std_error_df.unstack(1).transpose()

# Preview that Reformatting worked
transpose_df_tumor_response.head()
# transpose_df_tumor_response.columns
# transpose_df_std_error.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(timepoint,volume, xerr=std_error, yerr=mean_tumor_vol)

# Save the Figure
plt.savefig(r"Tumor Response to Treatment.png")
```

```python
# Show the Figure
plt.show()
```
